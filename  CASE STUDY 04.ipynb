# ● Check whether all the zones generate the same amount of sales.

Here we are using Anova test

```python
#Null Hypothesis,H0:All the zones generate the same amount of sales.
#Alt Hypothesis,Ha: All the zones  generate different  amount of sales.
```

```python
from scipy.stats import f_oneway
```

```python
 statistics,pvalue = f_oneway(data['Zone - A'], data['Zone - B'], data['Zone - C'],data['Zone - D'])
alpha = 0.05
print("The p value is",pvalue)
```

```python
if pvalue > alpha:
    print("Fail to reject the Null Hypothesis")
else:
    print ("Reject the Null Hypothesis")
```

since pvalue<alpha so "WE REJECT NULL HYPOTHESIS",which implies all the zones generate different amount of sales.
